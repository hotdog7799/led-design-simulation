```python
# 빈 row 생성 (컬럼 이름에 맞춰서 None 또는 원하는 기본값으로)
empty_row = pd.DataFrame([{col: None for col in yourmom.columns}])

# 빈 row 추가
yourmom = pd.concat([yourmom, empty_row], ignore_index=True)

# 다시 CSV로 저장
yourmom.to_csv("./simulation_log.csv", index=False)
```
